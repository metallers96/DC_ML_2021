Grouping by member_and_date

```python
ds=df.groupby(['Member_number','Date'])['itemDescription'].apply(lambda x: list(x))
```

```python
ds
```

```python
transactions = ds.values.tolist()
transactions[:10]
```

# Rules

```python
#applying apriori
from apyori import apriori
rules = apriori(transactions, min_support=0.00030,min_confidence = 0.05,min_lift = 2,min_length = 2)
results = list(rules)
results
```

```python
len(results)
```

```python
import os
```

```python
def print_rules(results):
    for rule in results:
        print ('rule.items=', list(rule.items))
        print ('rule.support=',rule.support)

        for os in rule.ordered_statistics:
            print ('\titems_base=', list(os.items_base))
            print ('\tlifted_item =', list(os.items_add))
            print ('\tlift=', os.lift)
            print ('\tconfidence (i.e. cond prob {} if {})='.format(list(os.items_add), list(os.items_base)), os.confidence)
            print ('\n')
        
print_rules(results)
```

```python
def print_rules_multi_os(results):
    for rule in results:
        if len (rule.ordered_statistics)>1:  
            print ('rule.items=', list(rule.items))
            for os in rule.ordered_statistics:
                print ('\n\titems_base=', list(os.items_base))
                print ('\tlifted_item =', list(os.items_add))
                print ('\tlift=', os.lift)
                print ('\tconfidence (i.e. cond prob {} if {})='.format(list(os.items_add), list(os.items_base)), os.confidence)

            break 

print_rules_multi_os(results)
```
